```python
import pygame
import time
import random

#for displaying the time of the text we need to initialize a font :
pygame.font.init()
```

```python
pygame.init()
```

```python
#now creating a window for our video game
WIDTH, HEIGHT = 1000, 600

# setting our video game background to a fixed width and height
WIN = pygame.display.set_mode((WIDTH, HEIGHT))

#setting the name on the top of our page or say captioning the page
pygame.display.set_caption("My Game")


# setting a background to our game 

BG = pygame.image.load("C:\\Users\\hp\\Downloads\\g_back.png")

```

```python
#now we have to create a game loop the loop which keeps the game running  
```

```python
#defining the size of the user
PLAYER_WIDTH = 40
PLAYER_HEIGHT = 40

#defining player_vel:
PLAYER_VEL = 6

FONT = pygame.font.SysFont("comicsans", 30)
Font = pygame.font.SysFont("comicsans", 60)


#width of the falling star:
STAR_WIDTH = 10
STAR_HEIGHT = 10

STAR_VEL = 4




def draw(player, elapsed_time, stars):    # a function to draw onto our window 
    WIN.blit(BG, (0, 0))   # blit the background into our game window

    time_text = FONT.render(f"Time: {round(elapsed_time)}s", 1, "white")
    WIN.blit(time_text, (10, 10))

    pygame.draw.rect(WIN, "red", player)

    for star in stars:
        pygame.draw.rect(WIN, "white", star)
    
    
    pygame.display.update()


#game logic function 
def main():

    run = True

    #creating a character/ object that can move around:
    # creating the shape of the user as a rectangle:

    player = pygame.Rect(200, HEIGHT - PLAYER_HEIGHT, PLAYER_WIDTH, PLAYER_HEIGHT)  #start the user block at the x coordinate 200 and the y coordinate as the bottom of the window
    clock = pygame.time.Clock()

    start_time = time.time()    # for the projectiles/ objects falling we have to keep note of the time when the game starts
    elapsed_time = 0


    #now we should add the projectiles in the form of increments
    star_add_increment = 1000 # after 2000milli seconds first star will fall
    star_count = 0   #variable which tells us when we should add an other star
    
    stars = [] # this is the list where we are storing all the stars currently on the screen and then we use thi sto draw them on the screen
    hit = False
    
    

    

    while run:

        clock.tick(60)   # it will set the while loop such that it will run only 60 times a second
        elapsed_time = time.time() - start_time   # calculating the current time with the help of time the game started

        #generating some stars
        star_count += clock.tick(60)
        #clock.tick returns the number of milliseconds since the last tick this is done to keep track of the precise time
        if star_count > star_add_increment:

            for i in range(3):
                star_x = random.randint(0, WIDTH - STAR_WIDTH)  # any random position on the window the star can be generated
                star = pygame.Rect(star_x, -STAR_HEIGHT, STAR_WIDTH, STAR_HEIGHT )   # the star starts above the top of the screen so -ve y coordinate 
                stars.append(star) # adding these three new stars into the stars list to display them on the screen

            star_add_increment = max(200, star_add_increment - 50)   # so the time in which the stars fall will be increaing as the time moves on
            star_count = 0
         
        
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:   # if they press the exit button the keyboard close the game
                run = False #there is no need to continue checking for events once the user press exit hence stop checking the conditions
                break 

        #designing the game moments: { based on the user keyboard actions}
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT] and player.x - PLAYER_VEL >= 0: #left arrow key code to set the block such that it doesnot go beyond the window
            player.x -= PLAYER_VEL # every time  the user preeses the left arrow button the block moves 5 units towards the (0 0) or say left
        if keys[pygame.K_RIGHT] and player.x + PLAYER_VEL + player.width <= WIDTH:
            player.x += PLAYER_VEL
            
            
        

        #after we design the stars and their moment and time now we have to draw them

        for star in stars[:]: # here we are making the list of the stars in the list because we can remove the stars that hit the bottom of the window
            star.y += STAR_VEL  # finding the y coordinate of the falling star

            if star.y > HEIGHT:
                stars.remove(star)

            elif star.y  + star.height >= player.y and star.colliderect(player):
                stars.remove(star)
                hit = True
                break

        # now setting the losing condition:

        if hit:
            lost_text = Font.render("YOU LOST!!!", 1, " Yellow")
            WIN.blit(lost_text, (WIDTH/2 - lost_text.get_width()/2, HEIGHT/2 - lost_text.get_height()/2))
            pygame.display.update()
            pygame.time.delay(7000)
            break
            
        draw(player, elapsed_time, stars)

         

    pygame.quit()

if __name__ == "__main__":
    main()

          
```
